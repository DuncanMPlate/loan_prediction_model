```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
X_train_dummies = pd.get_dummies(train_df)
X_test_dummies = pd.get_dummies(test_df)
X_test_dummies = X_test_dummies.reindex(labels=X_test_dummies.columns,axis=1)
```

```python
train_df
```

```python
test_df
```

```python
y_train_label = LabelEncoder().fit_transform(train_df['loan_status'])
y_test_label = LabelEncoder().fit_transform(test_df['loan_status'])
```

```python
X_train_dummies.columns
```

```python
X_test_dummies['debt_settlement_flag_Y'] = X_train_dummies['debt_settlement_flag_Y']
```

```python
# Convert categorical data to numeric and separate target feature for training data

X_train2019, X_test2019, y_train2019, y_test2019 = train_test_split(X_train_dummies, y_train_label)
```

```python
classifier = LogisticRegression()
classifier.fit(X_train2019, y_train2019)
print(f"Training Data Score: {classifier.score(X_train2019, y_train2019)}")
print(f"Testing Data Score: {classifier.score(X_test2019, y_test2019)}")
```

```python
X_train2020, X_test2020, y_train2020, y_test2020 = train_test_split(X_test_dummies, y_test_label)
```

```python
# Convert categorical data to numeric and separate target feature for testing data



classifier = LogisticRegression()
classifier.fit(X_test2020, y_test2020)
print(f"Training Data Score: {classifier.score(X_train2020, y_train2020)}")
print(f"Testing Data Score: {classifier.score(X_test2020, y_test2020)}")
```
